## Linear Regression

```python
class LinearRegressionNetwork(nn.Module):

  # Init the module architecture
  def __init__(self) -> None:
      super(LinearRegressionNetwork, self).__init__()

      self.linear_layer = nn.Linear(in_features=1,
                                    out_features=1)

  # Forward for an input x
  def forward(self, x):
    out = self.linear_layer(x)

    return out
```

```python
# Model Defiinition
model = LinearRegressionNetwork()

# Model to GPU
model = model.cuda()

print(f' > Model Architecture: \n\n{model}\n\n')
print(f' > Model Parameters: \n')
for key, value in model.state_dict().items():
  print(f'{key}: {value}')
```

## Classification

```python
class ClassificationNewtork(nn.Module):

  def __init__(self):
      super(ClassificationNewtork, self).__init__()
      self.linear_layer = nn.Linear(1,1)
      self.act_function = nn.Softmax() 
  
  def forward(self, x):
      z = self.linear_layer(x)
      out = self.act_function(z) 
      
      return out
```

```python
print('Classification\n\n')

# Model Defiinition
classification_model = ClassificationNewtork()

# Model to GPU
classification_model = classification_model.cuda()

print(f' > Model Architecture: \n\n{classification_model}\n\n')
print(f' > Model Parameters: \n')
for key, value in classification_model.state_dict().items():
  print(f'{key}: {value}')
```

# FeedForward Network

image.png

```python
class FeedforwardNetwork(nn.Module):
  def __init__(self, in_features, hidden_features):
      super(FeedforwardNetwork, self).__init__()

      # Input Layer 
      self.input_layer = nn.Linear(in_features, hidden_features)
      self.relu = nn.ReLU()

      # Hidden Layer
      self.hidden_linear = nn.Linear(hidden_features, 1)
      self.sigmoid = nn.Sigmoid()
  
  def forward(self, x):
      
      # Forward input layer
      h = self.input_linear(x)
      h = self.relu(h)

      # Forward hidden layer
      h = self.hidden_linear(h)
      out = self.sigmoid(h)
      
      return out
```

```python
print('FF network\n\n')

# Model Defiinition
in_features, hidden_features = 10, 5
ff_model = FeedforwardNetwork(in_features, hidden_features)

# Model to GPU
ff_model = ff_model.cuda()

print(f' > Model Architecture: \n\n{ff_model}\n\n')
print('-----------------------------\n')

print(f' > Model Parameters Shapes: \n')
for key, value in ff_model.state_dict().items():
  print(f'{key}:\t\t{value.shape}\n\n')
print('-----------------------------\n')

print(f' > Model Parameters Values: \n')
for key, value in ff_model.state_dict().items():
  print(f'{key}: {value}')
```
