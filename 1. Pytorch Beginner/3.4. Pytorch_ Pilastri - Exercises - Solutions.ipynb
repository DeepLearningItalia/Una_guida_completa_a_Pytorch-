### Solution

```python
import torch
from torch.utils.data import DataLoader
```

```python
from torchvision import datasets
from torchvision.transforms import ToTensor

training_data = datasets.CIFAR10(root='data',
                                 download=True,
                                 train=True,
                                 transform=ToTensor())

test_data = datasets.CIFAR10(root='data',
                             download=True,
                             train=False,
                             transform=ToTensor())
```

```python
# 1. Generate training and test dataloaders with a batch_size=4

batch_size = 4

train_dataloader = DataLoader(dataset=training_data,
                              batch_size=batch_size,
                              shuffle=True)

test_dataloader = DataLoader(dataset=test_data,
                             batch_size=batch_size,
                             shuffle=True)
```

```python
# 2. Find the dataset shape of training and test dataloaders

train_shape = train_dataloader.dataset.data.shape
print(f' 2. Training data shape: {train_shape}\n')

test_shape = test_dataloader.dataset.data.shape
print(f' 2. Test data shape: {test_shape}\n')
```

```python
# 3. Find the image shape of the dataloader

image_shape = train_dataloader.dataset.data[0].shape
print(f' 3. Image shape: {image_shape}\n')
```

```python
# 4. Plot one image from dataloader iteration
print('4. \n')
import matplotlib.pyplot as plt


for batch_image, batch_target in train_dataloader:
  print(f' > Batch Images shape: \t\t{batch_image.shape}\n')
  print(f' > Batch Targets: \t\t\t{batch_target} \t\t| shape: {batch_target.shape}\n')

  for image in batch_image:
    print(f' * Pay attention to this: each image has shape (3, x, y)')
    print(f' > Image shape: {image.shape}\n')
    image = image.transpose(0,2)    # TIP: Now the shape will be (y, x, 3)
    print(f' > Image shape after first transformation: \t\t\t{image.shape}')
    image = image.transpose(0,1)    # TIP: Now the shape will be (x, y, 3)
    print(f' > Image shape after second transformation: \t\t{image.shape}')

    plt.imshow(image)   # It needs an image of these shapes: (x,y,1) or (x,y,3)
    plt.show()

    break

  break
```

```python
# 5. Iterate over first 2 batches of training dataloader getting images, targets and classes
print(' 5. \n')
import matplotlib.pyplot as plt

loop = 2

i = 0
for batch_image, batch_target in train_dataloader:
  classes = [train_dataloader.dataset.classes[x] for x in batch_target]

  for idx, image in enumerate(batch_image):
    image = image.transpose(0,2)    # TIP: Now the shape will be (y, x, 3)
    image = image.transpose(0,1)    # TIP: Now the shape will be (x, y, 3)

    class_name = classes[idx]

    print(f'\n\n > Class: {class_name}')
    plt.imshow(image)   # It needs an image of these shapes: (x,y,1) or (x,y,3)
    plt.show()


  print('-------------------------')
  i += 1
  if i >= loop:
    break
```
